### Intro: find missing data

```python
import pandas as pd
import numpy as np
import math

crunchdf = pd.read_csv('crunchbase.csv')

crunchdf.columns.values
crunchdf
```

```python
# Here's building a function using def: 
def square_root(x): return math.sqrt(x)
```

```python
# Here's building the same function using lambda
sqrtL = lambda x: math.sqrt(x)

```

```python
# Use the isnull() method to detect the missing values. The output
# shows True when the value is missing. By adding an index into 
# the dataset, you obtain just the entries that are missing.
# A dataset could represent missing data in several ways. In this 
# example, you see missing data represented as np.NaN (NumPy Not 
# a Number) and the Python None value.

import re

def replace_non_digit(stringy):
    if stringy.__class__.__name__ == 'str':
        new_stringy = re.sub("[^0-9]", "",stringy)
        #For Alphabet things
        #new_stringy = re.sub("[^a-z]", "",stringy.lower())
        if not new_stringy:
            return 0
        return new_stringy
    return stringy

crunchdf[' funding_total_usd '] = [float(replace_non_digit(x)) for x in crunchdf[' funding_total_usd ']]

crunchdf
```

### Intro: fill in missing data

```python
# To fill in missing data use fillna(). For fillna() you need to 
# provide a number. Usually, the mean, median, or mode is used. 
# Let's use the same data set and this time let's fill in missing 
# values with the mean. 


#So there are other nan spots but this is the only column it makes sense
#to fill with the mean


#mean of non 0 values
total=[]
for x in crunchdf[' funding_total_usd ']:
    if x != 0.0:
        total.append(x)
funding_mean = np.mean(total)

#replace all 0's with the funding_mean
crunchdf.ix[(crunchdf[' funding_total_usd '] == 0.0), [' funding_total_usd ']] = funding_mean

#del crunchdf['Unnamed: 18']
crunchdf
```

```python
pd.pivot_table(crunchdf,index=["country_code", ' market ', "funding_rounds"], values = [' funding_total_usd '],aggfunc=np.sum).sort_values([' funding_total_usd '], axis=0, ascending = False)
```

```python
pd.pivot_table(crunchdf,index=[" market "], values = [' funding_total_usd '],aggfunc=np.sum).sort_values([' funding_total_usd '], axis=0, ascending = False)
```

```python
# We could also just drop all the NAs, by using dropna()
print crunchdf.dropna()
#What? This actually just gets rid of the ENTIRE column/row that has a single
# nan in it? How is that useful?
```
